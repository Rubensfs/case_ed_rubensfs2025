#Criação de estrutrara de diretorios e schemas

Criação de Diretorio para receber os arquivos

#Garantir que o schema raw existe no catalog datamasters

```python
%sql
CREATE SCHEMA IF NOT EXISTS datamasters.raw;
```

#Criar o VOLUME raw_inpe

```python
%sql
CREATE VOLUME IF NOT EXISTS datamasters.raw.raw_inpe;
```


#Criar o VOLUME raw_inpe

```python
%sql
CREATE VOLUME IF NOT EXISTS datamasters.raw.raw_tbra;
```

Verificar se o volume existe

```python
%sql
SHOW VOLUMES IN datamasters.raw;
```

#Criar os Schemas para a estrutura metalhão

#Bronze

```python
spark.sql("USE CATALOG datamasters")
```

```python
# 1️⃣ Cria o schema bronze inep dentro do catálogo atual (datamasters) 
spark.sql("CREATE SCHEMA IF NOT EXISTS b_inep")
```

```python
# 2️⃣ Define o schema como padrão dentro do catálogo
spark.sql("USE SCHEMA b_inep")
```

```python
spark.sql("SELECT current_catalog(), current_schema()").show()
```
