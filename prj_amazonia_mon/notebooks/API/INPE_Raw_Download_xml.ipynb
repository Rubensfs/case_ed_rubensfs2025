```python
# =========================================================
# Download XML TerraBrasilis (INPE) → Volume Databricks
# =========================================================

import requests
from datetime import datetime
```

```python
# ---------------------------------------------------------
# Configurações
# ---------------------------------------------------------
URL_XML = (
    "https://terrabrasilis.dpi.inpe.br/geonetwork/"
    "srv/api/records/fe02f2bf-2cc0-49d5-ab72-a3954f997408/formatters/xml"
)

VOLUME_PATH = "/Volumes/datamasters/raw/raw_tbra"

# Nome do arquivo com timestamp
data_ref = datetime.now().strftime("%Y%m%d_%H%M%S")
file_name = f"terrabrasilis_metadata_{data_ref}.xml"
file_path = f"{VOLUME_PATH}/{file_name}"
```

```python
# ---------------------------------------------------------
# Download do XML
# ---------------------------------------------------------
print("Iniciando download do XML TerraBrasilis...")

response = requests.get(URL_XML, timeout=60)
response.raise_for_status()

xml_content = response.text

print(f"Download concluído ({len(xml_content)} bytes).")
```

```python
# ---------------------------------------------------------
# Gravação no Volume (Unity Catalog)
# ---------------------------------------------------------
with open(file_path, "w", encoding="utf-8") as f:
    f.write(xml_content)

print(f"[✔] Arquivo gravado com sucesso em:")
print(file_path)
```
