```python
from pyspark.sql import functions as F
from datetime import date
```

```python
# Caminho base RAW
raw_path = "/Volumes/amazonia_catalog/raw/raw_inpe"
```

```python
# Data de carga (data do processamento)
data_ref_carga = date.today().strftime("%Y-%m-%d")
```

```python
# ==============================
# 1) CSVs DIÁRIOS
# ==============================
df_diario = (spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv(f"{raw_path}/focos_diario_br_*.csv")
    .withColumn("data_ref_carga", F.lit(data_ref_carga))
)

(df_diario.write
    .format("delta")
    .mode("append")
    .partitionBy("data_ref_carga")
    .saveAsTable("b_inep.d_foco_queim_inc")
)

print("✅ Tabela criada/atualizada: b_inep.d_foco_queim_inc")
```
