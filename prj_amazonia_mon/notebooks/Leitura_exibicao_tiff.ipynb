```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pyspark.sql import functions as F
```

```python
# Data de referência que você quer visualizar
referencia = "202509"
```

```python
# 1️⃣ Ler apenas os pixels da data de referência
df = spark.table("datamasters.b_inep.m_area_queim") \
          .filter(F.col("data_referencia") == referencia) \
          .select("linha", "coluna", "valor")
```

```python
# 2️⃣ Converter direto para Pandas (mais rápido para visualização)
pdf = df.toPandas()
```

```python
# 3️⃣ Reconstruir matriz 2D usando pivot
matriz = pdf.pivot(index='linha', columns='coluna', values='valor').values
```

```python
# 4️⃣ Exibir a imagem
plt.figure(figsize=(8,8))
plt.imshow(matriz, cmap='hot')  # cmap: 'hot', 'gray', 'viridis', etc
plt.title(f"Imagem de queimadas - {referencia}")
plt.colorbar(label="Valor de pixel")
plt.show()
```
